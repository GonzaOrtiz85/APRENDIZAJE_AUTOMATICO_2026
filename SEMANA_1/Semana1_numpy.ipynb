Ejercicio 7
Multiplicar dos matrices 3x3 de enteros aleatorios entre 1 y 10

```python
matriz1 = np.random.randint(1, 11, (3, 3))
matriz2 = np.random.randint(1, 11, (3, 3))

resultado = np.dot(matriz1, matriz2)

print("Matriz 1:")
print(matriz1)

print("Matriz 2:")
print(matriz2)

print("Multiplicación:")
print(resultado)
```

Ejercicio 8
Crear un array de 15 valores aleatorios entre 1 y 100 e imprimir máximo, mínimo y promedio

```python
array_random = np.random.randint(1, 101, 15)

print("Array:")
print(array_random)

print("Máximo:", np.max(array_random))
print("Mínimo:", np.min(array_random))
print("Promedio:", np.mean(array_random))
```

Ejercicio 9
Generar una matriz 5x5 con valores del 1 al 25 y extraer su diagonal

```python
matriz_1_25 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(matriz_1_25)

print("Matriz:")
print(matriz_1_25)

print("Diagonal:")
print(diagonal)
```

Ejercicio 10
Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcular la suma de cada fila y columna

```python
matriz_random = np.random.randint(1, 21, (5, 5))

suma_filas = np.sum(matriz_random, axis=1)
suma_columnas = np.sum(matriz_random, axis=0)

print("Matriz:")
print(matriz_random)

print("Suma de cada fila:")
print(suma_filas)

print("Suma de cada columna:")
print(suma_columnas)
```
