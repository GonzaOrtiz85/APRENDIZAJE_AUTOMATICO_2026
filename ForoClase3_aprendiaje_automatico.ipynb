```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")

print(df.columns)
print(df["origin"].unique())

df["origin_num"] = df["origin"].map({
    "usa": 1,
    "europe": 2,
    "japan": 3
})

plt.boxplot(
    [df["cylinders"].dropna(), df["origin_num"].dropna()],
    tick_labels=["Cilindros", "Origen"]
)

plt.title("Distribución de cilindros y origen")
plt.ylabel("Valores")
plt.grid(True)
plt.show()
```

```python
tabla = pd.crosstab(df["model_year"], df["origin"])

tabla.plot(kind="bar")
plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de autos")
plt.title("Autos por año y origen")
plt.legend(title="Origen")
plt.grid(True)
plt.show()
```
